```python
# Computation of mean and sum functions over sliding windows of 7 days each
window_length = 7
mean_over_windows = stock_pd.iloc[:, 1:3].rolling(window=window_length).mean()
sum_over_windows = stock_pd.iloc[:, 1:3].rolling(window=window_length).sum()

print(mean_over_windows)
```

```python
# Plotting the results
plt.figure(figsize=(10,5))
plt.plot(time_days, mean_over_windows , label=stock_pd.columns[1:3] )
plt.ylabel('stock price',fontsize=16)
plt.xlabel('date',fontsize=16)
plt.legend()
plt.title('Rolling window (mean)', fontsize=16)
plt.xticks(fontsize=14)
plt.show()


plt.figure(figsize=(10,5))
plt.plot(time_days, sum_over_windows , label=stock_pd.columns[1:3] )
plt.ylabel('stock price',fontsize=16)
plt.xlabel('date',fontsize=16)
plt.title('Rolling window (sum)', fontsize=16)
plt.legend()
plt.xticks(fontsize=14)
plt.show()
```

```python
# Implementation of non-overlapping window
def non_overlapped_window(x, win_size):
    n = len(x)
    return [x[i:i+win_size] for i in range(0, n, win_size)]

# Implementation of non-overlapping window with mean function
def non_overlapped_window_mean(x, win_size):
    n = len(x)
    return [np.mean(x[i:i+win_size]) for i in range(0, n, win_size)]

# Implementation of non-overlapping window with mean function
def non_overlapped_window_sum(x, win_size):
    n = len(x)
    return [np.sum(x[i:i+win_size]) for i in range(0, n, win_size)]
```

```python
# Computation of non-overlapping windows

new_tesla_stocks = non_overlapped_window(stock_pd['tesla_stocks'], 7)
new_apple_stocks = non_overlapped_window(stock_pd['apple_stocks'], 7)

print(new_tesla_stocks[:5])
```

```python
# Computation of mean and sum functions over non-overlapping windows

new_tesla_stocks_mean = non_overlapped_window_mean(stock_pd['tesla_stocks'], 7)
new_apple_stocks_mean = non_overlapped_window_mean(stock_pd['apple_stocks'], 7)

new_tesla_stocks_sum = non_overlapped_window_sum(stock_pd['tesla_stocks'], 7)
new_apple_stocks_sum = non_overlapped_window_sum(stock_pd['apple_stocks'], 7)
```
