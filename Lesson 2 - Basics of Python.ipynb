```python
#Check possible correlations among input parameters
import seaborn as sns

correlationMatrix = wine_dataset.corr()
print(correlationMatrix)
```

```python
# Plot correlation matrix in a better way
mask = np.triu(np.ones_like(correlationMatrix, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
colorPalette = sns.diverging_palette(240, 10)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap( correlationMatrix, annot=True, cmap=colorPalette, vmax=1, vmin=-1, center=0, linewidths=.15)
```

# **Exercise**

Plot all variables (one for each figure)

```python
 for i in range(len(column_names)):
    plt.figure()
    plt.plot(wine_dataset.iloc[:,i] )
    plt.title(column_names[i])
    plt.xlabel('sample',fontsize=16)
    plt.xticks(fontsize=14)
    plt.show()
```

# **Exercise**

Plot the histograms of all variables

```python
 for i in range(len(column_names)):
  plt.figure(figsize=(10,5))
  plt.hist(wine_dataset.iloc[:,i] , label=column_names[i] )
  plt.xlabel('value',fontsize=16)
  plt.xlabel('# of occurrencies',fontsize=16)
  plt.legend()
  plt.xticks(fontsize=14)
  plt.show()
```

# **Exercise**

Plot the free sulfur dioxide vs total sulfur dioxide (two highly correlated variables)


```python
plt.figure(figsize=(10,5))
plt.scatter(wine_dataset['total_sulfur_dioxide'] ,wine_dataset['free_sulfur_dioxide'] )
plt.xlabel('total sulfur dioxide',fontsize=16)
plt.ylabel('free sulfur dioxide',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()

```

```python
#histogram vs density plot
wine_dataset.hist( figsize=(20,10), bins=100 )
plt.show()

wine_dataset.plot(kind='density', subplots=True, layout=(4,3), sharex=False, figsize=(20,10) )
plt.show()
```
